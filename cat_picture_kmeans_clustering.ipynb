```python
import matplotlib as mpl
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
```

```python
# Import picture
image = mpl.image.imread('cat.jpg')
```

```python
plt.imshow(image)
```

```python
image.shape
```

```python
# Combine 2 axes

X = image.reshape(-1, 3)
X.shape
```

```python
# Cluster with KMeans

kmeans = KMeans(n_clusters=3, n_init=10)
```

```python
# Fit X 

kmeans.fit(X)
```

```python
segmented_image = kmeans.cluster_centers_[kmeans.labels_]
segmented_image = segmented_image.reshape(image.shape)
```

```python
plt.imshow(segmented_image/255)
```

```python
import cv2

cv2.imwrite('image1.png', cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
cv2.imwrite('image2.png', cv2.cvtColor(segmented_image.astype('uint8'), cv2.COLOR_BGR2RGB))
```
